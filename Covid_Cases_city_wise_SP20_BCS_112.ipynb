```
#import libraries
import folium
import pandas as pd
```

```
#create a map on Pakistan location
pakistan_map = folium.Map(location = [30.37,69.34],zoom_start=4.5,tiles='OpenStreetMap')
```

```
#import Google Drive and mount the entire drive
from google.colab import drive
drive.mount('/content/drive')
```

```
#load the dataset (csv file) into the Pandas dataframe
df = pd.read_csv('/content/drive/My Drive/IDS/CitiesLongLati.csv')
df
```

```
#calculate total number of cases using a loop
t = 0
for c in df['Total cases']:
  t = int(c)+int(t)
```

```
#mark circles on the map and displa the map
for District,lat,long,tc,ac,de,re in zip(list(df['District']),list(df['Latitudes']),list(df['Longitudes']),list(df['Total cases']),list(df['Active cases']),list(df['Deaths']),list(df['Recoveries'])):
    folium.CircleMarker(location = [lat,long],
                       radius = (tc/t)*100,
                       color = 'red',
                       fill = False,
                       fill_color = 'blue').add_to(pakistan_map)
    xyz = '<strong>District:  ' + District +'</strong><br>' + 'Total Cases: ' + str(tc)+'<br>' + 'Active cases: ' + str(ac)+ '<br>' + 'Deaths: ' + str(de)+ '<br>' + 'Recoveries: ' + str(re)
    iframe = folium.IFrame(xyz, width=200, height=150)
    popupData = folium.Popup(iframe, max_width=500)
    folium.Marker(location = [lat,long], popup=popupData).add_to(pakistan_map)
    

pakistan_map
```
